```python
import urllib.request,sys,time
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#code
pagesToGet= 238

upperframe=[]  
for page in range(2,pagesToGet+1):
    print('processing page :', page)
    url = 'https://www.politifact.com/article/list/?page='+str(page)
    print(url)
    
    #an exception might be thrown, so the code should be in a try-except block
    try:
        #use the browser to get the url. This is suspicious command that might blow up.
        page=requests.get(url)                             # this might throw an exception if something goes wrong.
    
    except Exception as e:                                   # this describes what to do if an exception is thrown
        error_type, error_obj, error_info = sys.exc_info()      # get the exception information
        print ('ERROR FOR LINK:',url)                          #print the link that cause the problem
        print (error_type, 'Line:', error_info.tb_lineno)     #print error info and line that threw the exception
        continue                                              #ignore this page. Abandon this and go back.
    time.sleep(1.25)   
    soup=BeautifulSoup(page.text,'html.parser')
    frame=[]
    links=soup.find_all('div',attrs={'class':'m-teaser'})
    print(len(links))
#     print(links)
    filename="PolitiFactNEWS.csv"
    f=open(filename,"w", encoding = 'utf-8')
    headers="Statement, Link, Date, Source \n"
    f.write(headers)
    cnt=0
    for j in links:
        Statement = j.find("h3",attrs={'class':'m-teaser__title'}).find('a').text.strip()
        Keywords = ["Russia", "Ukraine", "Ukrainian", "Russian"] 
        
        for k in Keywords:
            if k in Statement:
#                 print("he")
                cnt+=1
                Link = "https://www.politifact.com"
                Link += j.find("h3",attrs={'class':'m-teaser__title'}).find('a')['href'].strip()
                Date = j.find('div',attrs={'class':'m-teaser__meta'}).text[-19:-1].strip()
                Source = j.find('div', attrs={'class':'m-teaser__meta'}).text.strip()
                #Label = j.find('div', attrs ={'class':'m-teaser__image'}).find('img',attrs={'class':'c-image'}).get('alt').strip()
                frame.append((Statement,Link,Date,Source))
                f.write(Statement.replace(",","^")+","+Link+","+Date.replace(",","^")+","+Source.replace(",","^")+","+"\n")
                break
    if cnt>0:
        print(cnt)    
    upperframe.extend(frame)
f.close()
```

```python
data=pd.DataFrame(upperframe, columns=['Statement','Link','Date','Source'])
data.head()
```

```python
data.to_csv("PolitiFactNEWS.csv", index= False, encoding= 'utf-8')
df=pd.read_csv("PolitiFactNEWS.csv", encoding= 'utf-8')
df.head()
```
